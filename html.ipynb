```python
import os

# Create templates folder if not exists
os.makedirs("templates", exist_ok=True)
```

```python
html_code = """
<!DOCTYPE html>
<html>
<head>
    <title>Meeting Summary Generator</title>
    <style>
        body {
            background: #0f172a;
            color: white;
            font-family: Arial;
            padding: 40px;
        }
        textarea {
            width: 100%;
            height: 150px;
            padding: 10px;
        }
        button {
            padding: 10px 20px;
            margin-top: 10px;
            background: green;
            color: white;
            border: none;
            cursor: pointer;
        }
        #output {
            margin-top: 20px;
            background: #1e293b;
            padding: 20px;
        }
    </style>
</head>
<body>

<h1>Meeting Summary Generator</h1>

<textarea id="inputText" placeholder="Paste meeting text..."></textarea><br>

<button onclick="generate()">Generate Summary</button>

<div id="output"></div>

<script>
function generate() {
    let text = document.getElementById("inputText").value;

    fetch("/summarize", {
        method: "POST",
        headers: {
            "Content-Type": "application/json"
        },
        body: JSON.stringify({text: text})
    })
    .then(res => res.json())
    .then(data => {
        document.getElementById("output").innerHTML =
            "<h3>Summary:</h3>" + data.summary;
    });
}
</script>

</body>
</html>
"""
```

```python
# Save inside templates folder
with open("templates/index.html", "w") as f:
    f.write(html_code)

print("✅ index.html saved inside templates folder!")
```
